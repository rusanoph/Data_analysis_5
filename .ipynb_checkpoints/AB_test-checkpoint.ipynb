__Проверка типов данных__

В датасете logs_exp столбец EventTimestamp был переведен в формат datetime.

[Оглавление](#contents)

```python
def unix_to_datetime(unixtime):
    return datetime.utcfromtimestamp(unixtime).strftime('%Y-%m-%d %H:%M:%S')
# -------------------------------------------------------------------------------


df_logs_exp['EventTimestamp'] = list(map(unix_to_datetime, df_logs_exp['EventTimestamp']))
```

# 2. Проведение A/B тестирования

__Аггрегированные кумулятивные данные по дням о заказах__

Какой-то текст



```python
def agg_orders(x):
    group_by_date = np.logical_and(df_orders['date'] <= x['date'], df_orders['group'] == x['group'])

    agg_params = {'date' : 'max', 
                  'group' : 'max', 
                  'transactionId' : 'nunique', 
                  'visitorId' : 'nunique', 
                  'revenue' : 'sum'}

    return df_orders[group_by_date].agg(agg_params)
# -------------------------------------------------------------------------------


datesGroups = df_orders[['date','group']].drop_duplicates()


df_orders_aggregeted = datesGroups.apply(agg_orders, axis=1).sort_values(by=['date','group'])

display(df_orders_aggregeted)
```

__Аггрерированные кумулятивные данные по дням о посетителях__

```python
def agg_visitors(x):
    group_by_date = np.logical_and(df_visitors['date'] <= x['date'], df_visitors['group'] == x['group'])
    
    agg_params = {'date' : 'max', 
                  'group' : 'max', 
                  'visitors' : 'sum'}
    
    return df_visitors[group_by_date].agg(agg_params)
# -------------------------------------------------------------------------------


df_visitors_aggregeted = datesGroups.apply(agg_visitors, axis=1).sort_values(by=['date','group'])

display(df_visitors_aggregeted)
```

__Кумулятивные данные по дням о посетителях__

```python
cumulative_data = df_orders_aggregeted
cumulative_data['visitors'] = df_visitors_aggregeted['visitors']

cumulative_data
```

```python
cumulative_data_A = cumulative_data[cumulative_data['group'] == 'A']
cumulative_data_B = cumulative_data[cumulative_data['group'] == 'B']

# --Данные-для-построения--
cumulative_data_A_date = cumulative_data_A['date']
cumulative_data_A_revenue = cumulative_data_A['revenue']

cumulative_data_B_date = cumulative_data_B['date']
cumulative_data_B_revenue = cumulative_data_B['revenue']

# -------------------------------------------------------------------------------

plt.figure(figsize=(15, 9))
plt.plot(cumulative_data_A_date, cumulative_data_A_revenue, color='red')
plt.plot(cumulative_data_B_date, cumulative_data_B_revenue, color='blue')
plt.grid()
plt.title("График выручки по группам А и B")
plt.ylabel("Выручка")
plt.xlabel("Дата")

plt.tick_params(axis='x',rotation=45)
plt.legend(['A', 'B'], loc="upper left")

plt.show()
```

__Вывод__

А/B тестирование - сильный и эффективный инструмент развития бизнес-процессов, а также удобный критерий оценки для результатов эксперементальных разработок и внедрения изменений. Данный вид тестирования позволяет сравнить две версии чего-либо, чтобы проверить гипотезы и определить, какая версия продукта является более конкурентоспособной.

Было произведено A/B тестирование на наборе данных посетителей некоторого ресурса и их заказов. Были проанализированы изменения выручки от заказов в различных группах A и B.

По результатам теста было выявлено, что в группе B наблюдается увеличение выручки. Однако это изменение видно не сразу, скачок роста произошел спустя 2 недели после начала тестирования. Это говорит о том, что по мере внедрения изменений должно пройти достаточно времени, чтобы пользователи привыкли и адаптировались к изменению.
